```python
import numpy as np
import pandas as pd

import time
from datetime import datetime

from IPython.core.display import HTML
import tba3102



display(HTML("<style>pre { white-space: pre !important; }</style>"))
tba3102.set_default_pandas_options()

np.random.seed(int(round(time.time())))
random_state = 292652050 # np.random.randint(2**31-1)
print('random_state: {}'.format(random_state))
```

```python
print('Text processing started at {}'.format(datetime.now()))
```

```python
df = pd.read_csv('../data/training_1600000.csv', encoding='ISO-8859-1')
df = df.groupby('polarity').sample(frac=0.010, replace=False, random_state=random_state)
df.reset_index(inplace=True, drop=True)
df_without_emoticon = df_with_emoticon = df_with_emoticon_nospelling_nolemma = df
df
```

```python
df['polarity'].value_counts()
```

```python
tba3102.data_quality_report(df)
```

```python
stopword_exclusion_list = ['no', 'not']
tba3102.exclude_stopwords(stopword_exclusion_list)

df_without_emoticon = tba3102.normalize_corpus(df_without_emoticon, 'tweet', 'cleaned_tweet')
df_without_emoticon = df_without_emoticon.replace(r'^(\s?)+$', np.nan, regex=True)
df_without_emoticon = df_without_emoticon.dropna(subset=['cleaned_tweet']).reset_index(drop=True)

df_without_emoticon.to_csv('../data/training_16000_cleaned_without_emoticon.csv', index=False)
```

```python
df_with_emoticon = tba3102.normalize_corpus(df_with_emoticon, 'tweet', 'cleaned_tweet', 
                                            special_char_removal=False, remove_digits=False)
df_with_emoticon = df_with_emoticon.replace(r'^(\s?)+$', np.nan, regex=True)
df_with_emoticon = df_with_emoticon.dropna(subset=['cleaned_tweet']).reset_index(drop=True)

df_with_emoticon.to_csv('../data/training_16000_cleaned_with_emoticon.csv', index=False)
```

```python
df_with_emoticon_nospelling_nolemma = tba3102.normalize_corpus(df_with_emoticon_nospelling_nolemma, 'tweet', 'cleaned_tweet', 
                                                               special_char_removal=False, remove_digits=False, 
                                                               spelling_correction=False, lemmatize=False)
df_with_emoticon_nospelling_nolemma = df_with_emoticon_nospelling_nolemma.replace(r'^(\s?)+$', np.nan, regex=True)
df_with_emoticon_nospelling_nolemma = df_with_emoticon_nospelling_nolemma.dropna(subset=['cleaned_tweet']).reset_index(drop=True)

df_with_emoticon_nospelling_nolemma.to_csv('../data/training_16000_cleaned_with_emoticon_nospelling_nolemma.csv', index=False)
```

```python
print('Text processing ended at {}'.format(datetime.now()))
```
