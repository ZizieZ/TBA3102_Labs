```python
# Data Wrangling
import numpy as np
import pandas as pd
from collections import Counter

# Data Preparation
from sklearn.model_selection import train_test_split

# Feature Engineering
from sklearn.feature_extraction.text import CountVectorizer

# ML Models
from sklearn.tree import DecisionTreeClassifier

# Cross-Validation & Model Evaluation
from sklearn.model_selection import cross_val_score
import model_evaluation_utils as meu

# Visualization
from sklearn import tree
import graphviz
```

```python
raw_df = pd.read_csv('./data/sms_cleaned.csv')
raw_df.info()
```

```python
raw_df['Label'].unique()
```

## 2a. Preparing training & testing data with stratified sampling based on label with 80-20 ratio

```python
# Create label number
raw_df['Label_no'] = np.where(raw_df['Label'] == 'ham', 0, 1)
raw_df['Label_no'].unique()
```

```python
raw_df.sample(10)
```

```python
# Set random_state for reproducibility
random_state = 42
```

```python
x_train_corpus, y_test_corpus, x_train_label_nums, y_test_label_nums, x_train_label_names, y_test_label_names = train_test_split(np.array(raw_df['Cleaned_SMSText']),
                                                                                                                        np.array(raw_df['Label_no']),
                                                                                                                        np.array(raw_df['Label']),
                                                                                                                        test_size = 0.20, 
                                                                                                                        stratify = raw_df['Label_no'],
                                                                                                                        shuffle = True,
                                                                                                                        random_state = random_state)

print(f"Size of training corpus: {len(x_train_corpus)}\nSize of testing corpus: {len(y_test_corpus)}\nTotal: {len(raw_df)}")
```

```python
train_arr = np.bincount(x_train_label_nums)
test_arr = np.bincount(y_test_label_nums)
print(f"Size of ham sms in training: {train_arr[1]} and spam smses: {train_arr[0]}. Total: {len(x_train_label_nums)}\n"
     f"Size of ham smses in testing: {test_arr[1]} and spam smses: {test_arr[0]}. Total: {len(y_test_label_nums)}")
```

# 2b. Represent the unstructured SMS messages using bag of words model engineering the raw term occurrence features for the dataset

```python
# build BOW features on train articles
cv = CountVectorizer(binary = True, min_df = 0.0, max_df = 1.0) 
# Binary false: Returns Term Frequency
# Binary True: Returns occurence of term

# transform train & test data into features
cv_train_features = cv.fit_transform(x_train_corpus)
cv_test_features = cv.transform(y_test_corpus)

print('BOW model:> Train features shape:', cv_train_features.shape, ' Test features shape:', cv_test_features.shape)
```

## 2c. Train decision tree classifier to classify the SMSes & evaluate the performances with 5-fold cross validation

```python
# Initalize decision tree classifier
clf = DecisionTreeClassifier(criterion = 'gini', splitter = 'best', 
                             max_depth = 5, random_state = random_state)

# Fit training data to train model with 5-fold CV
clf.fit(cv_train_features, x_train_label_names)
clf_bow_cv_scores = cross_val_score(clf, cv_train_features, x_train_label_names, cv = 5)
print(f"Decision Tree Classifier (Training) Accuracy (5-fold): {clf_bow_cv_scores}\nDecision Tree Classifier (Training) mean training accuracy: {np.mean(clf_bow_cv_scores):.2f}")
```

```python
# Fit model to predict testing data
clf_predictions = clf.predict(cv_test_features)

# get unique labels
unique_classes = ['ham', 'spam']

# Get performance metrics
print(meu.get_metrics(true_labels = y_test_label_names, predicted_labels = clf_predictions))
```

```python
# Print confusion matrix
print(meu.display_classification_report(true_labels = y_test_label_names, predicted_labels = clf_predictions, classes = unique_classes))
```

## 2d. Generate Tree Visualization of the classifier & extract rules of classifying SMS

```python
dot_data = tree.export_graphviz(clf, max_depth = 5,                # Limit to a Depth of 5 only
                                feature_names = cv.get_feature_names_out(),
                                out_file = None, 
                                class_names = unique_classes,      # Provide Target Variable Column Name
                                filled = True, 
                                rounded = True,
                                special_characters=True)

graph = graphviz.Source(dot_data)  
graph 
```
