<b>1b. Represent review text data using VSM with minimal doc frequency of 0.01</b>

```python
# Create Tfidf Vectorizer object
tv = TfidfVectorizer(use_idf = True, min_df = 0.01)

# Engineer features with Tf-IDf 
tv_matrix = tv.fit_transform(sample_df['reviews.text_very_cleaned'])
```

```python
# Review DTM shape
print(f"TF-IDF's DTM shape: {tv_matrix.shape}")
```

```python
# Get all terms from the matrix
extracted_names = tv.get_feature_names_out()
len(extracted_names)
```

```python
# Get the features as an array
extracted_matrix = tv_matrix.toarray()

# Convert array with term as a DTM (dataframe)
tfidf_df = pd.DataFrame(extracted_matrix, columns = extracted_names)
tfidf_df.head()
```

```python
# Check number of terms extracted
tfidf_df.info() # 267 terms extracted based on minimal df of 0.01
```

```python
# Export tf-idf DTM as csv
tfidf_df.to_csv('./data/tfidf.csv')
```

<b>1c. Perform a pairwise simiarlity analysis on the TF-IDF features with consine similarity</b>

```python
similarity_matrix = cosine_similarity(tv_matrix)
similarity_df = pd.DataFrame(similarity_matrix)
```

```python
# Check similarity matrix
similarity_df.info()
```

```python
similarity_df.head(20)
```

```python
similarity_df.sample(5)
```

```python
# Helper: Identify highest pairwise document similarity
def extract_highest_sim(df, threshold = 0.0):
    """
    Arguments
    -----
    """
    # Append term result to list
    result = []
    results = []
    
    # iterate through the lower triangle of the matrix
    for i in range(len(df)):
        for c in range(i):
            if df.iloc[i,c] >= threshold and i != c:
                result.append(i)
                result.append(c)
                result.append(df.iloc[i,c])
                results.append(result)
                result = []
    
    return results
```

```python
print(sample_df['reviews.text_very_cleaned'].iloc[16330])
print(sample_df['reviews.text_very_cleaned'].iloc[6])
```

```python
# Helper: Sort and extract the highest terms from dataframe
def extract_terms(full_df, mat_df, threshold = 0.0, count = 1):
    """
    """
    # Extract unsorted list from df
    unsorted_lst = extract_highest_sim(mat_df, threshold)
    
    # Arrange list base on consine similarity
    sorted_lst = sorted(unsorted_lst, key = lambda x: x[2], reverse = True)
    
    # Prepare dataframe for results
    cols = ['doc1_index', 'doc1_review', 'doc1_cleanreview', 'doc2_index', 'doc2_review', 'doc2_cleanreview', 'cos_sim']
    
    # Prepare lists to get results
    doc1_index, doc1_review, doc1_cleanreview, doc2_index, doc2_review, doc2_cleanreview, cos_sim = [], [], [], [], [], [], []
    
    # Extract document(s) of highest similarity and place them side by side
    for i in range(count):
        doc1 = sorted_lst[i][0] # Document 1
        doc2 = sorted_lst[i][1] # Document 2
        sim = sorted_lst[i][2]  # Similarity between 2 documents
        
        doc1_index.append(doc1)
        doc1_review.append(full_df['reviews.text'].iloc[doc1])
        doc1_cleanreview.append(full_df['reviews.text_very_cleaned'].iloc[doc1])
        doc2_index.append(doc2)
        doc2_review.append(full_df['reviews.text'].iloc[doc2])
        doc2_cleanreview.append(full_df['reviews.text_very_cleaned'].iloc[doc2])
        cos_sim.append(sim)
    
    # Generate & return dataframe
    result_df = pd.DataFrame(list(zip(doc1_index, doc1_review, doc1_cleanreview, doc2_index, doc2_review, doc2_cleanreview, cos_sim)), columns = cols)
    return result_df
```

<b>1e. Create the flat clusters using the linkage matrix and maximum distance cutoff obtained from (1d)</b>

```python
# Set maximum distance cut-off
max_dist = 150

# Using fClusters from Scipy with distance as criterion
cluster_labels = fcluster(Z, max_dist, criterion = 'distance')
cluster_labels = pd.DataFrame(cluster_labels, columns = ['ClusterLabel'])
fcluster_df = pd.concat([sample_df.reset_index(drop=True), cluster_labels.reset_index(drop=True)], axis = 1)

fcluster_df.head()
```
